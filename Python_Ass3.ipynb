##### **Prb No : 1.1.** Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce().

```
from functools import reduce
```

```
# Solution 1

def myreduce(func, lst):
  if len(lst)==1:
    return lst[0]
  else:
    tempResult = func(lst[0], lst[1])
    for i in range(2,len(lst)):
      tempResult = func(tempResult, lst[i])
    return tempResult


add = lambda x,y:x+y
lst = [1,2,3,4,5]
result = myreduce(add,lst)
print(result)
```

```
# Solution 2

def myReduce(function, iterable):
  iterator = iter(iterable)
  if len(iterable)==1:
    return next(iterator)
  else:
    result = 0
    for i in iterator:
      result = function(result, i)
    return result

largestNumber = lambda x,y : x if x>y else y
lst = [1,2,3,4,5,6,30,45,15,100,80]
larNum = myReduce(largestNumber, lst)
print(larNum)
```

##### **Prb No : 1.2.** Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter().

```
# Solution 1

def myfilter(func, lst):
  evenLst = []
  for i in lst:
    if func(i):
      evenLst.append(i)
  return evenLst

evenNumbers = lambda x : x%2==0
lst = [1,2,3,4,5,6]
evenNums = myfilter(evenNumbers, lst)
print(evenNums)
```
