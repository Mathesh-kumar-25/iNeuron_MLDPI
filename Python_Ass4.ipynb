##### **Prb No : 1.1.** Write a Python Program(with class concepts) to find the area of the triangle using the below formula. 
##### **Formula** : area = (s*(s-a)*(s-b)*(s-c)) ** 0.5 
##### Function to take the length of the sides of triangle from user should be defined in the parent class and function to calculate the area should be defined in subclass.

```
# Solution 1 

class Shape:
  def __init__(self, s, a, b, c):
    self.s = s
    self.a = a
    self.b = b
    self.c = c

  def calculateArea(self):
    pass

class Triangle(Shape):
  def __init__(self, *args):
    super(Triangle, self).__init__(*args)

  def calculateArea(self):
    area = ((self.s)*(self.s-self.a)*(self.s-self.b)*(self.s-self.c))*0.5
    return area

  def __str__(self):
    return "Area of the triangle is {}".format(self.calculateArea())

triangle = Triangle(10,4,2,4)
print(triangle)
```

##### **Prb No : 1.2.** Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

```
# Solution 1

def filter_long_words(lst, n):
  wordsLonger = []
  for i in lst:
    if len(i)>n:
      wordsLonger.append(i)
  return wordsLonger

result = filter_long_words(['hi', 'hello', 'ineuron', 'MLDPI'], 5)
print(result)
```

##### **Prb No : 2.1.** Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words.
##### Hint: If a list [ ab,cde,erty] is passed on to the python function output should come as [2,3,4]. Here 2,3 and 4 are the lengths of the words in the list.

```
# Solution 1

def wordLength(lst):
  wordLen = []
  for i in lst:
    wordLen.append(len(i))
  return wordLen

result = wordLength(['hi', 'hello'])
print(result)
```

```
# Solution 2

def wordLength(lst):
    wordLen = list(map(len, lst)) 
    return wordLen

result = wordLength(['hi', 'hello', 'ineuron', 'MLDPI'])
print(result)
```

##### **Prb No : 2.2.** Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.

```
# Solution 1

def isVowel(chr):
  if chr.lower() in 'aeiou':
    return True
  else:
    return False

print("a is vowel", isVowel('a'))
print("E is vowel", isVowel('E'))
print("z is vowel", isVowel('z'))
```

```
# Solution 2

# Using lamda function
isvowel = lambda chr : True if chr.lower() in 'aeiou' else False

print("a is vowel", isvowel('a'))
print("E is vowel", isvowel('E'))
print("z is vowel", isvowel('z'))
```
