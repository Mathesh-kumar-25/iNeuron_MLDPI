##### **Prb No : 1.** Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```
# Solution 1

flag = 0
for i in range(2000, 3201):
  if (i%7==0) and (i%5!=0):
    if (flag==0):
      print(i, end="")
      flag = 1
    else:
      print(",", i, end="")
```

```
# Solution 2

numbersObtained = [i for i in range(2000, 3201) if (i%7==0) and (i%5!=0)]
print(",".join(str(i) for i in numbersObtained))
```

```
# Solution 3

startNumber = 0
for i in range(2000, 3201):
  if (i%7==0):
    startNumber = i
    break

numbersObtained = [i for i in range(startNumber, 3201, 7) if (i%5!=0)]
print(",".join(str(i) for i in numbersObtained))
```

##### **Prb No : 2.** Write a Python program to accept the user's first and last name and then getting them printed in the the reverse order with a space between first name and last name.

```
# Solution 1 (First & Last name reversed separately)

firstName = input("Enter first name : ")
lastName = input("Enter last name : ")
print(firstName[::-1], lastName[::-1])
```

```
# Solution 2 (First & Last name reversed in combined)

firstName = input("Enter first name : ")
lastName = input("Enter last name : ")
print(lastName[::-1], firstName[::-1])
```

##### **Prb No : 3.** Write a Python program to find the volume of a sphere with diameter 12 cm. (Formula: V=4/3 * Ï€ * r^3)

```
# Solution 1

diameterSphere = 12
radiusSphere = diameterSphere/2
volumeSphere = (4/3)*(3.14)*(radiusSphere**3)
print("Volume of Sphere is : ", volumeSphere)
```
