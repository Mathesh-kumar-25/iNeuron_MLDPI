##### **Prb No : 1.** Create the below pattern using nested for loop in Python.
pattern.png

```
# Solution 1

for i in range(5):
  for j in range(i+1):
    print('*',end="")
  print()

for i in range(5,1,-1):
  for j in range(i-1):
    print('*',end="")
  print()
```

```
# Solution 2

for i in range(1,6,1):
  print('*' * i)
for i in range(4,0,-1):
  print('*' * i)
```

##### **Prb No : 2.** Write a Python program to reverse a word after accepting the input from the user.

```
inputWord = input("Input word : ")
outputWord = inputWord[::-1]
print("Output word :",outputWord)
```
